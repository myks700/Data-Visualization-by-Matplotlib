### Create small multilpes plt.subplots

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(2,2)
ax[0,0].plot(st_weather["MONTH"], st_weather["MLY-PRCP-NORMAL"])
ax[0,1].plot(st_weather["MONTH"], st_weather["MLY-TAVG-NORMAL"])
ax[1,0].plot(aus_weather["MONTH"], aus_weather["MLY-PRCP-NORMAL"])
ax[1,1].plot(aus_weather["MONTH"], aus_weather["MLY-TAVG-NORMAL"])
plt.show()
```

```python
# Create a figure and an array of axes: 2 rows, 1 column with shared y axis
fig, ax = plt.subplots(2, 1, sharey=True)

# Plot Seattle precipitation data in the top axes
ax[0].plot(st_weather["MONTH"], st_weather["MLY-PRCP-NORMAL"], color = "b", marker = "*")
ax[0].plot(st_weather["MONTH"], st_weather["MLY-PRCP-25PCTL"], color = "b", linestyle = "-.")
ax[0].plot(st_weather["MONTH"], st_weather["MLY-PRCP-75PCTL"], color = "b", linestyle = "--")
plt.legend(["seattle_MLY-PRCP-NORMAL","MLY-PRCP-25PCTL", "MLY-PRCP-75PCTL"])

# Plot Austin precipitation data in the bottom axes
ax[1].plot(aus_weather["MONTH"], aus_weather["MLY-PRCP-NORMAL"], color = "r", marker = "*")
ax[1].plot(aus_weather["MONTH"], aus_weather["MLY-PRCP-25PCTL"], color = "r", linestyle = "-.")
ax[1].plot(aus_weather["MONTH"], aus_weather["MLY-PRCP-75PCTL"], color = "r", linestyle = "--")
plt.legend(["weather_MLY-PRCP-NORMAL","weather_MLY-PRCP-25PCTL", "weather_MLY-PRCP-75PCTL"], loc="upper left", bbox_to_anchor=(-0.2, -.15),
          ncol=3, fancybox=True, shadow=True)
plt.show()
```
